```python
# ── Cell 5: Write ocr.py (fixed) ────────────────────────────────────────────
with open('/content/cvip_project/pipeline/ocr.py', 'w') as f:
    f.write(r'''
import cv2
import numpy as np
import pytesseract
import fitz
from PIL import Image
from concurrent.futures import ThreadPoolExecutor, as_completed

pytesseract.pytesseract.tesseract_cmd = r'/usr/bin/tesseract'

def pdf_to_images(pdf_path):
    doc = fitz.open(pdf_path)
    pages = []
    for i, page in enumerate(doc):
        mat = fitz.Matrix(2.0, 2.0)
        pix = page.get_pixmap(matrix=mat)
        img = np.frombuffer(pix.samples, dtype=np.uint8).reshape(pix.h, pix.w, pix.n)
        if pix.n == 4:
            img = cv2.cvtColor(img, cv2.COLOR_RGBA2RGB)
        pages.append((i + 1, img))
    return pages

def preprocess_image(img):
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    binary = cv2.adaptiveThreshold(
        gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
        cv2.THRESH_BINARY, 31, 11
    )
    denoised = cv2.fastNlMeansDenoising(binary, h=10)
    return denoised

def _build_chunks(data, page_num, region_type):
    chunks = {}
    n = len(data["text"])
    for i in range(n):
        word = data["text"][i].strip()
        conf = int(data["conf"][i])
        block_num = data["block_num"][i]
        par_num = data["par_num"][i]
        line_num = data["line_num"][i]
        if not word or conf < 30:
            continue
        key = (page_num, block_num, par_num, line_num)
        if key not in chunks:
            chunks[key] = {
                "text": "",
                "page": page_num,
                "paragraph": par_num,
                "line": line_num,
                "block": block_num,
                "region_type": region_type,
                "confidences": []
            }
        chunks[key]["text"] += " " + word
        chunks[key]["confidences"].append(conf)
    result = []
    for chunk in chunks.values():
        chunk["text"] = chunk["text"].strip()
        chunk["avg_confidence"] = round(
            sum(chunk["confidences"]) / len(chunk["confidences"]), 1
        )
        del chunk["confidences"]
        if len(chunk["text"]) > 10:
            result.append(chunk)
    return result

def segment_text_regions(img, page_num):
    pil_img = Image.fromarray(img)
    data = pytesseract.image_to_data(pil_img, output_type=pytesseract.Output.DICT, config="--psm 3")
    return _build_chunks(data, page_num, region_type="text")

def segment_table_regions(img, page_num):
    pil_img = Image.fromarray(img)
    data = pytesseract.image_to_data(pil_img, output_type=pytesseract.Output.DICT, config="--psm 6")
    return _build_chunks(data, page_num, region_type="table")

def segment_figure_regions(img, page_num):
    _, thresh = cv2.threshold(img, 200, 255, cv2.THRESH_BINARY_INV)
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    chunks = []
    for i, cnt in enumerate(contours):
        x, y, w, h = cv2.boundingRect(cnt)
        if w * h < 10000:
            continue
        chunks.append({
            "text": "[FIGURE at T:%d B:%d L:%d R:%d]" % (y, y+h, x, x+w),
            "page": page_num,
            "paragraph": i,
            "line": 0,
            "block": i,
            "region_type": "figure",
            "avg_confidence": 100.0,
            "bounding_box": {"T": y, "B": y+h, "L": x, "R": x+w}
        })
    return chunks

def segment_font_regions(img, page_num):
    pil_img = Image.fromarray(img)
    data = pytesseract.image_to_data(pil_img, output_type=pytesseract.Output.DICT, config="--psm 7")
    return _build_chunks(data, page_num, region_type="font")

def segment_page_parallel(img, page_num):
    preprocessed = preprocess_image(img)
    workers = {
        "text":   (segment_text_regions,   preprocessed, page_num),
        "table":  (segment_table_regions,  preprocessed, page_num),
        "figure": (segment_figure_regions, preprocessed, page_num),
        "font":   (segment_font_regions,   preprocessed, page_num),
    }
    all_chunks = []
    with ThreadPoolExecutor(max_workers=4) as executor:
        futures = {
            executor.submit(fn, img_arg, pg): name
            for name, (fn, img_arg, pg) in workers.items()
        }
        for future in as_completed(futures):
            worker_name = futures[future]
            try:
                all_chunks.extend(future.result())
            except Exception as e:
                print("[WARNING] Worker %s failed on page %d: %s" % (worker_name, page_num, str(e)))
    return all_chunks

def process_pdf(pdf_path):
    all_chunks = []
    pages = pdf_to_images(pdf_path)
    print("Processing %d pages with parallel segmentation..." % len(pages))
    for page_num, img in pages:
        print("  Page %d/%d..." % (page_num, len(pages)), end="\r")
        all_chunks.extend(segment_page_parallel(img, page_num))
    print("\nDone. Total chunks: %d" % len(all_chunks))
    return all_chunks
''')

print("ocr.py written successfully!")
```

```python
rag_code = '''
import chromadb
from chromadb.utils import embedding_functions
from pipeline.ocr import process_pdf

CHROMA_PATH = "./chroma_db"

def get_collection():
    client = chromadb.PersistentClient(path=CHROMA_PATH)
    ef = embedding_functions.SentenceTransformerEmbeddingFunction(
        model_name="all-MiniLM-L6-v2"
    )
    return client.get_or_create_collection(
        name="cvip_books",
        embedding_function=ef
    )

def index_pdf(pdf_path: str, book_name: str) -> int:
    chunks = process_pdf(pdf_path)
    collection = get_collection()

    # Remove figure chunks — not searchable text
    chunks = [c for c in chunks if c.get("region_type") != "figure"]

    documents, metadatas, ids = [], [], []
    seen_ids = set()  # track duplicates

    for chunk in chunks:
        doc_id = (
            f"{book_name}"
            f"_p{chunk['page']}"
            f"_b{chunk['block']}"
            f"_par{chunk['paragraph']}"
            f"_ln{chunk.get('line', 0)}"
            f"_{chunk['region_type']}"
        )

        # Skip duplicate IDs — keeps first occurrence only
        if doc_id in seen_ids:
            continue
        seen_ids.add(doc_id)

        documents.append(chunk["text"])
        metadatas.append({
            "book": book_name,
            "page": chunk["page"],
            "paragraph": chunk["paragraph"],
            "line": chunk.get("line", 0),
            "block": chunk["block"],
            "region_type": chunk.get("region_type", "text"),
            "confidence": chunk["avg_confidence"]
        })
        ids.append(doc_id)

    batch_size = 100
    for i in range(0, len(documents), batch_size):
        collection.upsert(
            documents=documents[i:i+batch_size],
            metadatas=metadatas[i:i+batch_size],
            ids=ids[i:i+batch_size]
        )

    return len(ids)  # actual unique chunks indexed

def search(query: str, n_results: int = 5) -> list:
    collection = get_collection()

    # Safety check — if collection empty
    if collection.count() == 0:
        return []

    results = collection.query(
        query_texts=[query],
        n_results=min(n_results, collection.count())
    )

    output = []
    for i, doc in enumerate(results["documents"][0]):
        distance = results["distances"][0][i]
        # Skip irrelevant results
        if distance > 1.2:
            continue
        output.append({
            "text": doc,
            "metadata": results["metadatas"][0][i],
            "distance": round(distance, 4)
        })
    return output
'''

with open('/content/cvip_project/pipeline/rag.py', 'w') as f:
    f.write(rag_code)

print("rag.py ")
```

```python
# ── Cell 7: Write agents.py  ─────────────────────────────────────────
with open('/content/cvip_project/pipeline/agents.py', 'w') as f:
    f.write(r'''
from langchain_ollama import OllamaLLM
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from pipeline.rag import search
import re

llm = OllamaLLM(model="gemma2:2b", temperature=0)
parser = StrOutputParser()

def _is_valid_chunk_fast(chunk):
    text = chunk["text"]
    conf = chunk.get("avg_confidence", 0)
    if conf < 40:
        return False
    if len(text) < 10:
        return False
    alpha = sum(c.isalpha() for c in text)
    if alpha / len(text) < 0.4:
        return False
    return True

def agent1_validate_chunk(chunk):
    conf = chunk.get("avg_confidence", 0)
    if conf >= 70 and len(chunk["text"]) > 20:
        chunk["valid"] = True
        chunk["agent1_verdict"] = "VALID (fast)"
        return chunk
    if not _is_valid_chunk_fast(chunk):
        chunk["valid"] = False
        chunk["agent1_verdict"] = "INVALID (fast)"
        return chunk
    prompt = PromptTemplate(
        input_variables=["text", "confidence"],
        template='You are a document quality checker.\nDecide if this OCR text is VALID (real readable text) or INVALID (noise/symbols/garbage).\nText: "{text}"\nOCR Confidence: {confidence}%\nReply with exactly one word: VALID or INVALID'
    )
    chain = prompt | llm | parser
    result = chain.invoke({"text": chunk["text"], "confidence": conf}).strip()
    chunk["valid"] = result.upper() == "VALID"
    chunk["agent1_verdict"] = result
    return chunk

def agent2_correct_text(chunk):
    if chunk.get("avg_confidence", 100) >= 75:
        chunk["corrected_text"] = chunk["text"]
        chunk["agent2_corrected"] = False
        return chunk
    prompt = PromptTemplate(
        input_variables=["text"],
        template='You are an OCR post-correction expert.\nFix ONLY clear OCR errors (l->1, O->0, broken hyphens). Do NOT rewrite.\nReturn only the corrected text.\nOCR text: "{text}"\nCorrected text:'
    )
    chain = prompt | llm | parser
    chunk["corrected_text"] = chain.invoke({"text": chunk["text"]}).strip()
    chunk["agent2_corrected"] = True
    return chunk

def agent3_query(user_query):
    results = search(user_query, n_results=5)
    if not results:
        return "No matching content found in the indexed books."
    context_parts = []
    for r in results:
        m = r["metadata"]
        part = "[Book: %s | Page: %s | Paragraph: %s | Line: %s]\n%s" % (
            m["book"], m["page"], m["paragraph"], m.get("line", "N/A"), r["text"]
        )
        context_parts.append(part)
    context = "\n\n".join(context_parts)
    prompt = PromptTemplate(
        input_variables=["query", "context"],
        template='You are a document search assistant.\nUser searched for: "{query}"\n\nRelevant passages found:\n{context}\n\nTell the user: book name, page number, paragraph number, line number, and a brief summary.\nList all locations if multiple found. Be precise and concise.'
    )
    chain = prompt | llm | parser
    return chain.invoke({"query": user_query, "context": context}).strip()

def run_full_pipeline(chunks):
    validated = [agent1_validate_chunk(c) for c in chunks]
    valid_only = [c for c in validated if c.get("valid", False)]
    corrected = [agent2_correct_text(c) for c in valid_only]
    return corrected
''')

print("agents.py written successfully!")
```

```python
# ── Cell 8: Write app.py with Gradio ────────────────────────────────────────
app_code = '''
import gradio as gr
import os
import sys
sys.path.insert(0, "/content/cvip_project")

from pipeline.ocr import process_pdf
from pipeline.rag import index_pdf, search
from pipeline.agents import agent3_query

# ── Upload + Index ────────────────────────────────────────────────────────────
def process_and_index(pdf_file):
    if pdf_file is None:
        return "No file uploaded."

    save_path = os.path.join("/content/cvip_project/uploads", os.path.basename(pdf_file.name))
    with open(pdf_file.name, "rb") as src, open(save_path, "wb") as dst:
        dst.write(src.read())

    book_name = os.path.basename(pdf_file.name).replace(".pdf", "")

    # OCR
    chunks = process_pdf(save_path)
    if not chunks:
        return "No text extracted from PDF."

    # Skip agents for speed — direct index
    for c in chunks:
        c["text"] = c.get("corrected_text", c["text"])

    count = index_pdf(save_path, book_name)

    # Preview first 3 chunks
    preview = ""
    for i, c in enumerate(chunks[:3]):
        preview += f"\\n--- Chunk {i+1} ---\\n"
        preview += f"Page: {c['page']} | Para: {c['paragraph']} | Type: {c.get('region_type','text')}\\n"
        preview += f"Text: {c['text'][:150]}...\\n"

    return f" Indexed {count} unique chunks from '{book_name}'\\n\\nPreview:{preview}"


# ── Search ────────────────────────────────────────────────────────────────────
def search_books(query, use_agent, n_results):
    if not query.strip():
        return "Please enter a search query.", ""

    raw = search(query, n_results=int(n_results))

    if not raw:
        return "No relevant matches found. Try a different query.", ""

    # Raw results
    raw_output = ""
    for i, r in enumerate(raw):
        m = r["metadata"]
        raw_output += f"\\n{'='*40}\\n"
        raw_output += f"Match {i+1}\\n"
        raw_output += f"Book: {m['book']}\\n"
        raw_output += f"Page: {m['page']} | Paragraph: {m['paragraph']} | Line: {m.get('line', 'N/A')}\\n"
        raw_output += f"Type: {m.get('region_type', 'text')} | Confidence: {m['confidence']}%\\n"
        raw_output += f"Distance: {r['distance']}\\n"
        raw_output += f"Text: {r['text']}\\n"

    # Agent 3 answer
    agent_output = ""
    if use_agent:
        agent_output = agent3_query(query)

    return raw_output, agent_output


# ── Gradio UI ─────────────────────────────────────────────────────────────────
with gr.Blocks(title="CVIP Book Search") as demo:

    gr.Markdown("# CVIP — Book Search System")
    gr.Markdown("Upload a PDF book, then search for any text to get exact page, paragraph and line numbers.")

    with gr.Tab("Upload & Index"):
        gr.Markdown("### Upload a PDF book")
        pdf_input = gr.File(label="Choose PDF", file_types=[".pdf"])
        index_btn = gr.Button("Process & Index", variant="primary")
        index_output = gr.Textbox(label="Result", lines=12, interactive=False)
        index_btn.click(fn=process_and_index, inputs=[pdf_input], outputs=[index_output])

    with gr.Tab("Search"):
        gr.Markdown("### Search the book archive")
        query_input = gr.Textbox(label="Search text", placeholder="e.g. CVIP is TOP")
        with gr.Row():
            use_agent = gr.Checkbox(label="Use Agent 3 (LLM answer)", value=False)
            n_results = gr.Slider(minimum=1, maximum=10, value=5, step=1, label="Number of results")
        search_btn = gr.Button("Search", variant="primary")
        with gr.Row():
            raw_output = gr.Textbox(label="Raw matches", lines=15, interactive=False)
            agent_output = gr.Textbox(label="Agent 3 Answer", lines=15, interactive=False)
        search_btn.click(
            fn=search_books,
            inputs=[query_input, use_agent, n_results],
            outputs=[raw_output, agent_output]
        )

demo.launch(share=True, debug=False)
'''

with open('/content/cvip_project/app.py', 'w') as f:
    f.write(app_code)

print("app.py updated with Gradio!")
```
